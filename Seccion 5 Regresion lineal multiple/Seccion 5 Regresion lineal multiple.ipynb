## Idea de la Regresion Lineal Multiple

Para ello tomaremos en parte esa formula 

$y=b_0+(b_1 \cdot X_1)+(b_2\cdot X_2)+...+(b_n\cdot X_n)$

Siendo $y$ la variable dependiente, aquella que queremos predecir en nuestro caso los ingresos que va a generar la empresa 

$b_0$ siendo la constante 

$b_n$ siendo los coefisientes

$X_n$ siendo las variables independientes

En este caso haremos lo sig

$y$ ganacias

$b_0$ Ganancia base si no invertimos en marketing

$b_1 \cdot X_1$ cuanto por cada unidad de gasto en I+D ganamos mas o disminuimos provecho

$b_2 \cdot X_2$ gasto de asministracion

$b_3 \cdot X_3$ gasto adicional por marketing

### Restricciones en modelor de regresion lineal

# Construccion de modelo paso a paso

A compacacion de la Regresion lineal simple en la cual contabamos con una variable independiente para predecir su variable dependiente pero la Regresion lineal multiple tiene mas de dos vasiables independientes que quieeren ser potenciales predictores de la variable dependiente 
 por lo tanto nuestro trabajo sera empezar a decidir cuales queremos conservar de modo que un modelo que prediga menos varibales es mas preferible que el original 

no podemos conservar el modelo original con todas las variables por dos motivos

1. no por añadir mas variables vamos a tener mas informacion, sera un modelo que estara lleno de basura y que no por tener mas variables hara una mejor prediccion
2. si el numero de variables crece y crece para predecir la variable dependiente hara que el dia que se tenga que explicar el funcionamiento del modelo sera mas complejo y desgastante. Por tanto la idea es mantener solo las variables importantes y las que son capaces de predecir algo

Exsisten varios modos de construitr modelos

1. Exautivo (All-in) \\
Termino que viene del casino cuando uno lo apuesta todo. Es decir que meteremos todas las variables al modelo. Este tipo de modelos es mas usado si hay alguno de estos puntos:
 * Conocimiento a priori; OR \\
 Si tenemos un conocimiento previo y ya sabemos que variables son exactamente las que usaremos, todas son las que tienen que aparecer en el modelo.
 * Por nesecidad; OR
 * Preparacion previa para Eliminacion hacia atras \\
 Se genera un modelo con todas las variables y apartir de alli se hace la tecnica
2. Regresion gradual o Regresion paso a paso \\
Se basa en ir construyendo un modelo y actualizarlo añadiendo o quitando una variable

 2. Seleccion hacia adelante: aunque parezca que sera identico al anterior solo que al revez es todo lo contratio, entraremos en el modelo, entraremos en un modelo sin variables e iremos añadiendo aquellas que tengan un p-valor inferior el nivel de significacion \\

## Construccion del modelo RLM usando Eliminacion hacia atras

```python
import statsmodels.api as sm
```

no tenemos de ningun modo reflejado el termino independiente o la ordenada en el origen entonces al evaluar la significatividad estadistica del modelo podria ser que lo que sobra fuera el termino independiente 

Pero no tenemos ningun modo de checarlo automaticamente, entonces lo que hacemos generalmente agregar una columna vertical al conjunto de datos correspondida de 1´s y se entiende que el coeficiente que se lleva los 1´s precisamente sera nuestro termino independiente

```python
X= np.append(arr=np.ones((50,1)).astype(int), values=X, axis=1)
#insertaremos en una sola columna 50 filas de unos y axis nos ayudara a indicar que lo añadiremos a columna (1)
#en este caso agregara primero X y despues los unos
#si deseamos al revez debemos primero agregar X en arr y depsues los unos en values
```

```python
X
```

Usaremos la variable `X_opt` para almacenar el num optimo de variables independientes estadisticamente significativas para ser capaces de predecir la variable dependiente

```python
X_opt = X[:, [0,1,2,3,4,5]].tolist() #recordemos que la eliminacion hacia atras comienza con todas las variables y se quedara con las que sirven
```

```python
SL = 0.05 #crearemos el nivel de significacion
```

crearemos el modelo para nuestra regresion llamado OLS (ordinary list of squares) o tecnica de los cuadrados ordinarios

```python
regression_OLS = sm.OLS(endog=y, exog=X_opt).fit()
```

Siendo `endog` la variable que queremos predecir `exog` representa la matriz de caracteristicas y con `.fit()` justaremos el modelo

Consultaremos el P-valor de cada variable y observar cual es el mas grande y supera S.L. para eliminarlo 

Para hacerlo aplicaremos la funcion `.summary()` a `regression_OLS ` esto devolvera un P-valor tambien los coeficientes del modelo y parametros

```python
regression_OLS.summary()
```

es esta tabla nos indica
* **coef** el coeficiente de cada una de las variables del modelo sindo const el sueldo base, x1 y x2 las variables dummy, x3 el gasto en i+d, x4 el gasto en administracion y x5 el gasto en marketing 
* **std err** el error estandar de la variable 
* **t** el estadistico t
* **P>|t|** el P-valor de cada variable
* los 2 sig. son los intervalos de confianza

Una idea general por el momento es que el P-valor es una probailidad y cuanto mas baja es ese P-valor mas estadisticamente significativa es esa variable

Por ejemplo, en este modelo x3 que es el gasto i+d y cont que es el termino independiente tiene un P-valor de 0.000 por tanto son las mas estadisticamente significativas, mientras el resto son poco relevantes o poco estadisticamente significativas con respecto siempre a la variable dependiente que intentamos predecir que es el ingreso de la empresa

ahora el siguente paso es observar cual tiene mas pvalor conforme si nuestro SL es mayor, por lo tanto eliminaremos x2




```python
X_opt_1 = X[:, [0,1,3,4,5]].tolist()
regression_OLS_1 = sm.OLS(endog=y, exog=X_opt_1).fit()
regression_OLS_1.summary()
```

```python
X_opt_2 = X[:, [0,3,4,5]].tolist()
regression_OLS_2 = sm.OLS(endog=y, exog=X_opt_2).fit()
regression_OLS_2.summary()
```

## Eliminación hacia atrás automática
Si estás interesado en implementaciones automáticas de la Eliminación hacia atrás en Python, aquí te presentamos dos de ellas. Se ha adaptado el código para que utilice la transformación `.tolist()` sobre el ndarray y así se adapte a Python 3.7.

### Eliminación hacia atrás utilizando solamente p-valores:

```python
import statsmodels.formula.api as sm
def backwardElimination(x, sl):    
    numVars = len(x[0])    
    for i in range(0, numVars):        
        regressor_OLS = sm.OLS(y, x.tolist()).fit()        
        maxVar = max(regressor_OLS.pvalues).astype(float)        
        if maxVar > sl:            
            for j in range(0, numVars - i):                
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):                    
                    x = np.delete(x, j, 1)    
    regressor_OLS.summary()    
    return x 
 
SL = 0.05
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
X_Modeled = backwardElimination(X_opt, SL)
```
