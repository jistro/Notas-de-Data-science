# Que son
Las maquinas de soporte vectorial sirven tanto para regresiones lineales como no lineales, para ello van a depender de un kernel, el cual determina el tipo de regresión que llevara a cabo por eso se llaman de soporte vectorial

La idea será ajustar el mayor corredor posible, en lugar de ajustar el mayor corredor (o la mayor calle) posible entre dos clases, ajustando el margen de los mismos como lo hacen las Maquinas de Soporte Vectorial, las Maquinas de Soporte de Regresión intentando mantener cuantas más observaciones posibles del conjunto de datos dentro del corredor en torno a la recta limitando unos márgenes máximos 

La anchura del pasillo (o calle) en torno a la recta se controla mediante un hiper parámetro $\epsilon$ (épsilon) 

La SVR lleva a cabo una regresión lineal en un espació de dimensión superior, es decir en lugar de hacerlo en el plano, va a hacerlo en un espacio de dimensión muy elevada 



## SVR

Podemos pensar que al crear una SVR es como si cada punto de datos de l conjunto de entrenamiento representara su propia dimensión del problema, cuando evaluamos el núcleo o kernel entre un punto de test y uno de training, el resultado es la coordenada del punto de test de dicha dimensión 

image.png


Se necesitará que el conjunto de entrenamiento $T= \{ \vec{X},\vec{Y} \}$ cubra el dominio de interés y vaya acompañado de las soluciones en dicho dominio

El trabajo de la SVM es aproximar la función que se ha usado para generar los puntos del conjunto de entrenamiento: $F(\vec{X}) = \vec{Y}$

En un problema de clasificación, los vectores X se usan para definir un hiperplano que separe las dos categorías en nuestra solución

Estos vectores se usan para llevar a cabo la regresión lineal. Los vectores más cercanos al punto de prueba se llaman vectores de soporte. Podemos evaluar nuestra función en cualquier lugar, por lo que cualquier vector podría estar mas cerca de nuestra ubicación de evaluación de prueba

# Construir un modelo SVR
1. Tener un conjunto de entrenamiento, $T=\{ \vec{X},\vec{Y} \}$
2. Elegir un núcleo y sus parámetros, así como llevar a cabo cualquier regularización que sea necesaria
3. Crear matriz de correlaciones, $K$
4. Entrenar el modelo, de forma exacta o aproximada para obtener los coeficientes de contracción, $\vec{\alpha}=\{ \alpha_{i} \}$
5. Utilizar estos coeficientes parar crear un estimador 


$f(\vec{X},\vec{\alpha},x^{*})=y$

## Elegir el kernel
Elegir el kernel 
* Kernel lineal, podemos imaginar que es como el que se usa para la regresión lineal 
$\langle x,y \rangle$ </center>
* Kernel no lineal $\langle \varphi (x),\varphi(y) \rangle = K(x,y)$
El mas usado es el gaussiano 

Luego debemos ver si necesitamos algo de regularización para eliminar o no el ruido 

$K_{i,j}=exp(\sum_{k}\theta_k | x_{k}^{i} -x_{k}^{j}| ^{2})+\varepsilon \delta_{i,j}$

La parte principal del algoritmo es $K\vec\alpha = \vec y $
* $\vec y$ es el vector de valores del conjunto de entrenamiento 
* $K$ Es la matriz de correlaciones 
* $\vec\alpha$ Es el conjunto de incógnitas para las cuales resolvemos el sistema de ecuaciones lineal como
$\vec\alpha = K^{-1} \vec y$
* Cuando tenemos la estimación de los parametros $\vec{\alpha}$ usamos los coeficientes que hemos calculado durante el paso de optimizacion y el kernel que habíamos al principio
* Para estimar el valor de $y^{*}$ para un punto de test $x^{*}$ calculamos el vector de correlaciones $\vec{k}$, $K_{i,j}=exp(\sum_{k}\theta_k | x_{k}^{i} -x_{k}^{j}| ^{2})$ de donde: $y^{*} = \vec{\alpha}\cdot\vec{k}$
